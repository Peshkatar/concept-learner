```python
# data transformation
def df_transformation(df_):
    return (
        df_
         .astype(
             {"capital_run_length_longest": np.int16, 
              "capital_run_length_total": np.int16, 
              "is_spam": np.int8}))
```

```python
# all but the last three columns (continuous values)
def df_discretization(df_):
    return (
        pd
        .DataFrame(KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform').fit_transform(df_), 
                   columns=columns[:-3], 
                   dtype=np.int8))
```

```python
df_transformed = df_transformation(df.iloc[:, -3:])
df_discretized = df_discretization(df.iloc[:, :-3]) 
```

```python
# assert that changes in data size didnt effect the values in the table
assert (df.iloc[:, -3:].describe() == df_transformed.describe()).any().any()
```

```python
integrated_df = df_discretized.join(df_transformed)
integrated_df.head()
```
