```python
def discretization(df_: pd.DataFrame, nbins: int, columns_: list[str], class_column: str) -> pd.DataFrame:
    """discretizes dataframe into n_bins, joins the class column post-discretization and lastly converts the table data type into 8-bit integer."""
    return (
        pd
        .DataFrame(
            KBinsDiscretizer(n_bins=nbins, encode='ordinal', strategy='quantile')
            .fit_transform(df_), 
            columns=columns_)
        .join(df[class_column])
        .astype(np.int8))
```

### Concept-learning algorithm implementation

```python
def LGG_set(D: pd.DataFrame) -> pd.DataFrame:
    """
    1 2 3 4 5 6 7
    Input : data D.
    Output : logical expression H . 
    1. x â† first instance from D; 
    2. H â† x;
    3. while instances left do
    4. x â†next instance from D;
    5. H â†LGG(H,x); 
    6. end
    7. return H
    """
    D = D.loc[D.is_spam == 1, :]
    H = D.iloc[[0], :] # first row
    
    for i, row in D.iloc[:, :-1].iterrows():
        if i < D.shape[0]-1:
            x = D.iloc[[i+1], :] # next row in iteration
            H = LGG_conj(H, x)
    return H
```

```python
def LGG_conj(H: pd.DataFrame, x: pd.DataFrame) -> pd.DataFrame:
    """
    Input : conjunctions H, x.
    Output : conjunction z.
    1. z â† conjunction of all literals common to x and y;
    2. return z;
    """
    w = H.to_dict(orient="record")[0]
    y = x.to_dict(orient="record")[0]
    
    return pd.DataFrame([{key:value for key, value in w.items() if w[key] == y[key]}])
```
